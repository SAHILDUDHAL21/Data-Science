# A) Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data (Use iris.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('iris.csv')
sc=df['variety'].value_counts()
plt.pie(sc,labels=sc.index)
plt.axis('equal')
plt.title('PIE PLOT OF FREQUENTIES OF THE THREE SCIPES OF IRIS DATA')
plt.show()
```
